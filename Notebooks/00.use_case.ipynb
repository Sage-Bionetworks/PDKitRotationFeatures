### Get Required Libraries

```python
from PDKitRotationFeatures import gait_module  
import pandas as pd
import numpy as np
import synapseclient as sc
%load_ext autoreload
%autoreload 2
```

### Instantiate Object

```python
syn = sc.login()
featureObjs = gait_module.GaitFeaturize()
```

### Demo

```python
## get sample data
data = syn.tableQuery("SELECT * FROM syn12514611 where recordId = '5bed85d5-a088-4ffd-8820-af56f36c3cb6'")
file_dict = syn.downloadTableColumns(data, columns = ["walk_motion.json"])
ts = pd.read_json(file_dict["42317198"])

## get data into right format
ts_accel_sample = ts[ts["sensorType"] == "userAcceleration"].rename(
    {"timestamp":"t"}, axis = 1)[["t", "x", "y", "z"]].reset_index(drop = True)
ts_gyro_sample = ts[ts["sensorType"] == "rotationRate"].rename(
    {"timestamp":"t"}, axis = 1)[["t", "x", "y", "z"]].reset_index(drop = True)
```

### Run Feature Pipeline

```python
result = featureObjs.run_gait_feature_pipeline(ts_accel_sample, ts_gyro_sample)
result.head(10)
```
